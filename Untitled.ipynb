### Split the columns of interest into attributes and labels, as shown in the tutorial. Then, split that data into a train (80%) and test set (20%). 

### Train a linear regression model on the data, as shown in the tutorial. Print the y-intercept, and the slope given by the model. Include this in your report. (make sure you also have the code for this).

```python
X = dataset['MinTemp'].values.reshape(-1,1)
y = dataset['MaxTemp'].values.reshape(-1,1)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)


regressor = LinearRegression()  
regressor.fit(X_train, y_train) #training the algorithm

#To retrieve the intercept:
print(regressor.intercept_)
#For retrieving the slope:
print(regressor.coef_)
```

### As in the tutorial, construct 2 columns in a new dataframe named “Actual” and “Predicted”. Print this dataframe, copy and paste it in your report.

### Add a third Column to your new dataframe called “Error”. Compute the error (absolute difference) between the actual value and the predicted value, and fill the new column with those values. Print the dataframe, copy and paste it in your report.

```python
y_pred = regressor.predict(X_test)
df = pd.DataFrame({'Actual': y_test.flatten(), 'Predicted': y_pred.flatten()})
df

df['Error'] = pd.Series( abs(df.Actual-df.Predicted) ,index=df.index)
print("\n"+str(df.head()))
```

### What is the mean of the error? What is the standard deviation of your error? Print these values and copy and paste it in your report.

```python
print("\nMean for error column is :"+str(df['Error'].mean()))
print("\nStandard Deviation for error column is :"+str(df['Error'].std()))
```

### By binning, compute the mean and std of the error of your model for each separate bin (range of values, in degrees). That is, what is the mean/std of the error of your model for each bin, where a bin is 10 degrees. So, you need a mean error in the interval (-40 to -30), another from (-30 to -20), (-20 to -10), and so on. 

```python
bins = [0, 10,20, 30, 40,50,60]
df['binned'] = pd.cut(df['Error'], bins)

mean=(df.groupby(['binned'])['Error'].mean())
print(mean)
std=(df.groupby(['binned'])['Error'].std(ddof=0))
print(std)
```
